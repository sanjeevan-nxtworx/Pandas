###Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

###Line Plot (Trend Over Time)
**A line plot is useful for showing trends over time, such as monthly sales data.**

```python
# Sample Sales Data
sales = pd.Series(
    [10500, 9800, 11200, 13450, 8900, 14300, 15600, 13400, 9800, 11000, 10200, 12350],
    index=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
)

# Plot the line graph
plt.figure(figsize=(10, 5))
plt.plot(sales, marker='o', linestyle='-', color='b', label="Sales")

# Customizing the plot
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.xticks(rotation=45)  # Rotate x-axis labels
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

###Bar Chart (Comparing Categories)
**A bar chart is useful for comparing categories (e.g., sales in each month).**

```python
plt.figure(figsize=(10, 5))
plt.bar(sales.index, sales, color="skyblue")

plt.title("Monthly Sales Bar Chart")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')

plt.show()
```

###Histogram (Distribution of Data)
**A histogram helps visualize the distribution of sales data.**

```python
plt.figure(figsize=(8, 5))
plt.hist(sales, bins=5, color="orange", edgecolor="black")

plt.title("Sales Distribution")
plt.xlabel("Sales Amount ($)")
plt.ylabel("Frequency")
plt.grid(True)

plt.show()
```

###Pie Chart (Proportion of Data)
**A pie chart helps show percentage contribution of different months to total sales**

```python
plt.figure(figsize=(7, 7))
plt.pie(sales, labels=sales.index, autopct='%1.1f%%', colors=["red", "green", "blue", "orange", "purple", "yellow", "pink", "cyan", "brown", "gray", "lime", "magenta"])

plt.title("Sales Distribution by Month")
plt.show()
```

###Box Plot (Outliers & Distribution)
**A box plot helps detect outliers in the data.**

```python
import seaborn as sns

plt.figure(figsize=(5, 5))
sns.boxplot(y=sales, color="lightblue")

plt.title("Sales Distribution - Box Plot")
plt.ylabel("Sales ($)")
plt.grid(True)

plt.show()
```

###Seaborn Bar Plot
**Seaborn provides a better-styled bar chart.**

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 5))
sns.barplot(x=sales.index, y=sales, palette="coolwarm")

plt.title("Sales per Month - Seaborn")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.xticks(rotation=45)

plt.show()
```

###Aggregation and Grouping

##Generate a DataFrame for the same


```python
import numpy as np
from datetime import datetime, timedelta
np.random.seed(42)
num_rows = 100

dates = [datetime(2024, 1, 1) + timedelta(days=np.random.randint(0, 30)) for _ in range(num_rows)]
regions = np.random.choice(['North', 'South', 'East', 'West'], num_rows)
products = np.random.choice(['Product A', 'Product B', 'Product C', 'Product D'], num_rows)
sales = np.random.randint(100, 1000, num_rows)
quantity = np.random.randint(1, 20, num_rows)
```

```python
data = {
    'Date': dates,
    'Region': regions,
    'Product': products,
    'Sales': sales,
    'Quantity': quantity
}

df = pd.DataFrame(data)
```

```python
df
```

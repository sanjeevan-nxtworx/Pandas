### Using Matplot Lib

###Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

###**Line Plot (Trend Over Time)**

```python
#A line plot is useful for showing trends over time, such as monthly sales data.

# Sample Sales Data
sales = pd.Series(
    [10500, 9800, 11200, 13450, 8900, 14300, 15600, 13400, 9800, 11000, 10200, 12350],
    index=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
)
```

```python
#Set figure size
plt.figure(figsize=(10, 5))

# plt.figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None, frameon=True, clear=False, layout=None)
# Parameter	Description
# num	        -   (Optional) Integer or string. Specifies the figure number or name.
# figsize	    -   (Optional) Tuple (width, height) in inches. Controls figure size.
# dpi	        -   (Optional) Dots per inch (resolution). Default is 100.
# facecolor	  -   (Optional) Background color of the figure.
# edgecolor	  -   (Optional) Edge color of the figure.
# frameon	    -   (Optional) If False, removes the figure frame. Default is True.
# clear	      -   (Optional) If True, clears existing figure before creating a new one.
# layout	    -   (Optional, Matplotlib 3.4+) Automatic layout management ('tight', etc.).
```

```python
plt.plot(sales, marker='o', linestyle='-', color='b', label="Sales")
```

###Creating Multiple Figures

```python
plt.figure(num=1)  # First figure
plt.plot([1, 2, 3], [4, 5, 6], color='red', label="Figure 1")
plt.legend()
plt.title("First Figure")

plt.figure(num=2)  # Second figure
plt.plot([1, 2, 3], [10, 20, 30], color='blue', label="Figure 2")
plt.legend()
plt.title("Second Figure")

plt.show()
```

###Plot Properties

```python
#These control lines, markers, colors, and styles.
# Property	        Description
# color='r'	      Sets the color of the plot (e.g., 'b', 'g', 'r', etc.).
# linestyle='--'	Sets the line style (e.g., '-', '--', ':', '-.').
# linewidth=2	    Sets line thickness.
# marker='o'	    Sets marker style (e.g., 'o', 's', '^').
# markersize=8	  Sets marker size.
```

```python
plt.plot([1, 2, 3], [4, 5, 6], color='r', linestyle='-.', marker='s', markersize=8, linewidth=1)
plt.show()
```

###Labels & Titles
Control text elements of the plot.

**Property**  
plt.title("My Chart")	    -- Sets the plot title.

plt.xlabel("X Axis")	    -- Labels the x-axis.

plt.ylabel("Y Axis")	-- Labels the y-axis.

plt.xticks(rotation=45)	-- Rotates x-axis labels.


```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("My Line Chart")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
plt.plot(sales, marker='o', linestyle='-', color='b', label="Sales")

# Customizing the plot
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.xticks(rotation=45)  # Rotate x-axis labels
plt.grid(True)
plt.legend()


```
